```python
import requests as rs
from bs4 import BeautifulSoup as bs
import time
domain = "https://travelking.taipei"
res = rs.get('https://travelking.taipei/attractions/')
soup = bs(res.text)
fid = open("python/travelking_taipei.txt","w")
for list in soup.select('.masonry_list .link')[2:-1]:
    url = domain + list['href']
    place_res = rs.get(url)
    place_soup = bs(place_res.text)
    if place_soup.select('#po span'):
        a = str(place_soup.select('#po span')[0].text.encode('utf-8'))
    if place_soup.select('span[itemprop="address"]'):
        b = str(place_soup.select('span[itemprop="address"]')[0].text.encode('utf-8').split('台北市')[-1].split('區')[0] + "區")
    if place_soup.select('span[itemprop="address"]'):
        c = str(place_soup.select('span[itemprop="address"]')[0].text.encode('utf-8'))
    if place_soup.select('#map_img a img'):
        d = str(place_soup.select('#map_img a img')[0]['src'].split('scene.png|')[-1].encode('utf-8').split(',')[0])
        e = str(place_soup.select('#map_img a img')[0]['src'].split('scene.png|')[-1].split(',')[1].split('&sensor=')[0])
    if place_soup.select('#point a'):
        f = str(place_soup.select('#point a')[0].text.encode('utf-8').strip())
    txt = a + "," + b + "," + c + "," + d + "," + e + "," + f
    print txt
    fid.write(txt)
    fid.write('\n')
    time.sleep(1)
fid.close()
```
